# **Case1:The risk is measured by the total number of infections in a country.**

```python
df['risk'] = df['Total Confirmed'].apply(lambda x: 0 if x<df['Total Confirmed'].mean() else 1)
x = df[['Adult.smoking', 'Median Earnings 2010', 'Violent.crime','vaccination_percentage_12+','median_age','female_percentage','high school diploma rate','Total Hospitals (2019)','Total Primary Care Physicians (2019)','Total Specialist Physicians (2019)','Median_Household_Income_2018','Unemployment_rate_2018']]
y = df['risk']
x = np.array(x)
y = np.array(y)

X_train, X_test, y_train, y_test= train_test_split(
    x, y,test_size=0.2, random_state=0)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler() 

sc.fit(X_train)

print(sc.scale_, sc.mean_)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
X_train_std[0]
```

```python
df['risk'].value_counts()
```

```python
t=0
for i in y_test:
  if i==0: t=t+1
print(t)
```

```python
from sklearn import metrics
from sklearn.metrics import roc_curve, auc, roc_auc_score
def train_and_get_importance(model, model_name):
    print(model_name+'----traning')
    model.fit(X_train_std,y_train)
    y_pred = model.predict(X_test_std)
    print('Misclassified samples: %d' % (y_test != y_pred).sum())
    from sklearn.metrics import accuracy_score
    print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
    print(metrics.classification_report(y_test, y_pred, target_names=['0', '1']))
    con_mat=metrics.confusion_matrix(y_test, y_pred)
    print(con_mat)
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
    print ('auc score', auc(false_positive_rate, true_positive_rate))
    try:
        feature = model.feature_importances_
        print('feature importances are:')
        feature_name = ['Adult.smoking', 'Median Earnings 2010', 'Violent.crime','vaccination_percentage_12+','median_age','female_percentage','high school diploma rate','Total Hospitals (2019)','Total Primary Care Physicians (2019)','Total Specialist Physicians (2019)','Median_Household_Income_2018','Unemployment_rate_2018']
        for i in  range(len(feature)):
            print(feature_name[i],':',feature[i])
        most_impotant = feature_name[np.argmax(feature)]
        print('for {}, the most important feature is "{}"'.format(model_name,  most_impotant))
    except:
        pass
    print('------------------------------------------------------------------------')
    
for model, model_name in zip(models, model_names):
    train_and_get_importance(model, model_name)
```
